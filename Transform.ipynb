```python
#transform on the wrong date
#1. load the data
import pandas as pd

df = pd.read_csv('Loan_Data.csv')
print(df)
```

```python
#2. exploration for wrong datafor col in df.columns:
unique_cols = df[col].unique()
print(f"The unique values in every column are : \n{unique_cols}")
```

```python
#to change the wrong data on dependent column
print(df[df['Dependents']=='3+'])
print(df[df['Dependents']=='3+']['Loan_ID'])
```

```python
df.replace({'Dependents': {'3+':3}},inplace = True)
print(df['Dependents'].unique())
```

```python
#2.missing data
#missing data exploration
df.isnull().sum()
```

```python
#Gender,Married,Dependents,Self_Employed are the columns with missing values that can be altered - these are discrete values.
#adding the missing values - since these are categorical, we are going with mode()[0]
df.fillna({
    'Gender':df['Gender'].mode()[0],
    'Married':df['Married'].mode()[0],
    'Dependents':df['Dependents'].mode()[0],
    'Self_Employed':df['Self_Employed'].mode()[0]
},inplace = True)
print(df)
print(df.isnull().sum())
print(df['Gender'].unique())
print(df['Married'].unique())
print(df['Dependents'].unique())
print(df['Self_Employed'].unique())
```

```python
df.dropna(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Dependents'] =  df['Dependents'].astype('int64')
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].astype('int')
df.dtypes
```

```python
df.drop(columns = ['Loan_ID'],inplace = True)
```

```python
#dummy encoding or one hot encoding for the categorical columns
binary_cols = ['Gender','Married','Education','Loan_Status','Self_Employed','Credit_History']
df[binary_cols] = pd.get_dummies(df[binary_cols],drop_first = True, dtype = int)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Property_Area'] = le.fit_transform(df['Property_Area'])

df
```

```python
df.to_csv('Cleaned_csv', index = False)
```
